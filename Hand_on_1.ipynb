```python
# 1.X Gate (NOT gate)
qc_x = QuantumCircuit(1)
qc_x.x(0)
qc_x.measure_all()
qc_x.draw('mpl')
```

```python
# 2. H Gate
qc_h = QuantumCircuit(1)
qc_h.h(0)
qc_h.measure_all()
qc_h.draw('mpl')
```

```python
# 3. Z Gate
qc_z = QuantumCircuit(1)
qc_z.z(0)
qc_z.measure_all()
qc_z.draw('mpl')
```

```python
# 4.CNOT Gate
qc_cx = QuantumCircuit(2)
qc_cx.cx(0,1)
qc_cx.measure_all()
qc_cx.draw('mpl')
```

```python
# 5.Bell State creation
qc_bell = QuantumCircuit(2)
qc_bell.h(0)
qc_bell.cx(0,1)
qc_bell.measure_all()
qc_bell.draw('mpl')
```
